##### Task
#####  You are asked to create a customer and order table as follows "Customers" and "Orders". Determine the right datatype to create the two tables. Here are the structures and data for both tables:4

#####  Once you have successfully created both tables, create a query to retrieve the total number of orders made by each customer along with the customer name from the "Customers" and "Orders" tables. Expected Output:



```sql

-- Membuat table customers
create table customers (
    customer_id serial,
    customer_name varchar(50),
    city varchar(50),
    primary key (customer_id)
);

-- Membuat table orders
create table orders (
	order_id serial,
	customer_id integer,
	order_date date,
	total_amount float,
	primary key (order_id)
);

-- Memasukkan data ke dalam table customers
insert into customers (customer_name, city) 
values ('John Doe', 'New York'), ('Jane Smith', 'Los Angeles'), ('David Johnson', 'Chicago');

-- Memasukkan data ke dalam table orders
insert into orders (customer_id, order_date, total_amount) 
values (1, '2022-01-10', 100.00), (1, '2022-02-15', 150.00), (2, '2022-03-20', 200.00), (3, '2022-04-25', 50.00); 

-- Melakukan join antara table customers dan orders dengan table customer_id, kemudian select customer_name dan menghitung menggunakan aggregation function count untuk melihat total order yang dilakukan oleh customer
select c.customer_name, count(o.customer_id) as total_orders
from customers as c join orders as o on o.customer_id = c.customer_id
group by c.customer_name order by 2 desc;
